# Selecting

**iloc -> Depreciated method. Used to select data from rows and columns using numbers**
<br />
**loc -> Used to select data from rows and columns using the column's/row's name**
<br />
**These are not functions but rather properties of the dataframe**

### Selecting from rows

```python
n = 1
df.iloc[[n]] # Get the data from row n

# Multiple rows can be printed by providing a starting and ending index
# df.iloc[x:y]
# x = starting row index | y = ending row index. The element at this index is not included
df.iloc[0:4]
# Here the data from the 0th to 3rd rows are printed

# iloc also supports comma separated index positions for rows 
df.iloc[[1,2]]
# Here, the 1st and 2nd rows are printed
```

### Selecting from columns

```python
n = 1
df.iloc[:,[n]] # Get the data from column n

# Multiple columns can be printed by providing a starting and ending index
# df.iloc[:,x:y]
# x = starting column index | y = ending column index. The element at this index is not included
df.iloc[:,0:4]
# Here the data from the 0th to 3rd columns are printed

# iloc also supports comma separated index positions for columns 
df.iloc[:,[1,2]]
# Here, the 1st and 2nd columns are printed
```

### Selecting from rows and columns

```python
# Can print data from rows and columns like a mini dataframe
# df.iloc[x:y,a:b]
# x = starting row index | y = ending row index. The element at this index is not included
# a = starting column index | y = ending column index. The element at this index is not included
# Note: The first index for row/column is always 0
df.iloc[2:4,1:2]
# Here, we are selecting the rows from index 2 to 4, so rows 2,3 are included
# For the column we are selecting from index 1 to 4, so only the column at 1st index is included

# iloc also supports comma separated index positions for rows and columns 
df.iloc[[1,2],[0,2,4]]
# Here, the 1st and 2nd rows and the 0th, 2nd and 4th columns are printed
```

### Selecting with loc function

```python
# Using loc, we can selects rows and columns based on their names

# With regards to rows, loc works the same way as iloc. However, loc includes the data at the ending index when selecting multiple rows
df.loc[0:4]
# Prints data from 0th row to 4th row

# With regards to columns, loc can be used to specify a particular column to select from/multiple columns

df.loc[:,["START*"]] # 1 column

df.loc[:,"START*":"STOP*"] # Multiple columns via range

df.loc[:,["START*","CATEGORY*","STOP*"]] # Multiple columns via commas
```

```python
df.loc[0:4,["START*","STOP*"]]
```

# Note

`df.iloc[n]` or `df.loc[n]` - returns a series and it is the raw data in the column
<br />
`df.iloc[[n]]` or `df.loc[[n]]` - returns the column in the dataframe format i.e. a neat and clean format


# Filtering

Extracting data which meets certain condition(s)

```python
df[df["MILES*"] > 100] # Prints the rows where the value in the miles column is greater than 100
df.loc[df["MILES*"] > 150,["START_DATE*","START*","STOP*","MILES*"]] # Prints the start date, start, stop and miles columns of the rows where the value in the miles column is greater than 150
```

```python
df.loc[df["STOP*"] == "New York"] # Prints the rows where the value in the stop column is New York
df.loc[df["START*"].isin(["New York","Jamaica","Cary"])] # Prints the rows where the value in the start column is present in the specified list
```

### Combining conditional expressions with logical operators

```python
df.loc[(df["MILES*"] > 10) & (df["START*"].isin(["New York","Cary"]))] # Using and logical operator

df.loc[(df["START*"] == "New York") | (df["STOP*"] == "New York")] # Using or logical operator
```
